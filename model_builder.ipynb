```python
import pickle
import pandas as pd
pkl_filename = "Dataset.pkl"
with open(pkl_filename, 'rb') as file:
    df=pd.read_pickle(file)
```

```python
df
```

```python

num_classes = len(df['Sub_Category_3'].unique())
num_classes
label_map = {label: i for i, label in enumerate(df['Sub_Category_3'].unique())}
int_labels = [label_map[label] for label in df['Sub_Category_3'].unique()]
```

```python
def encodeer_classes(x):
    return(label_map[x])
```

```python


df['Sub-Category-3_encoded'] = df['Sub_Category_3'].apply(lambda x: encodeer_classes(x))

import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout,Reshape, LSTM
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split


# Split your data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df['vector'], df['Sub-Category-3_encoded'], test_size=.25, random_state=104, shuffle=True)
x_train, X_test, Y_train, y_test = train_test_split(df['vector'], df['Sub-Category-3_encoded'], test_size=.7, random_state=104, shuffle=True)

# Convert your vectorized text data into numpy arrays
X_train = np.array(list(X_train))
X_test = np.array(list(X_test))



# Convert your labels to one-hot encoded vectors
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)


# Define your model architecture
model = Sequential()
model.add(Dense(512, input_dim=1024, activation='relu'))
model.add(Dropout(0.2))
model.add(Reshape((2, 256))) 
model.add(Dropout(0.2))
model.add(LSTM(128))

model.add(Dropout(0.3))
model.add(Dense(num_classes, activation='softmax'))

# Compile your model
from keras.optimizers import SGD
opt = SGD(lr=0.01)

model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])

# Train your model
model.fit(X_train, y_train, epochs=600, batch_size=20, validation_data=(X_test, y_test))

# Evaluate your model
loss, accuracy = model.evaluate(X_test, y_test)
print("Loss:", loss)
print("Accuracy:", accuracy)
```

```python
model.save('./sub_category_3_model/')
```

```python
import pickle
pkl_filename = "sub_category_3_labelmap.pkl"
with open(pkl_filename, 'wb') as file:
    pickle.dump(label_map, file)
```
